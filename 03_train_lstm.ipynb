# 03_train_lstm

## 목적
- 순차 패턴 기반 모델인 `LSTM` 실험을 별도로 정리한다.
- 주문 간격의 **순서 정보**가 집계 특징 대비 추가 성능을 주는지 확인한다.

## 입력 파일
- `01_prepare_data.ipynb`에서 생성한 sequence input
- 같은 사용자 기준 train / val / test split

## 출력 파일
- 권장: `results/test_predictions_lstm.csv`
- 권장: `results/thresholds.json`

## 핵심 역할
- 고객별 과거 주문 간격 시퀀스를 시간 순서대로 인코딩한다.
- 마지막 hidden state 또는 pooled representation으로 지연 위험 score를 계산한다.
- 집계 특징 모델과 동일한 라벨 정의를 유지한 채 순차 정보의 추가 가치를 점검한다.

## 공정성 체크
1. `target_gap > 15 -> delay_risk = 1` 규칙은 tabular 모델과 동일해야 한다.
2. 사용자 split은 `02_train_mlp.ipynb`와 완전히 동일해야 한다.
3. decision threshold는 validation set에서 따로 튜닝하되, test set에서는 다시 조정하지 않는다.
