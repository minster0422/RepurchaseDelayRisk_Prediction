# 03_train_lstm

## 목적
- LSTM을 순차 패턴 기반 모델로 정의하고, MLP와의 비교 의미를 분명히 한다.
- 핵심 질문은 **순차 정보가 실제 예측에 도움이 되는가**다.

## 입력 파일
- `01_prepare_data.ipynb`에서 생성해야 할 sequence 입력
- 같은 사용자 기준 train / val / test split

## 출력 파일
- 권장: `results/test_predictions_lstm.csv`
- 권장: `results/thresholds.json`

## 현재 상태
- 현재는 **설계 및 가이드 수준의 노트북**이다.
- 저장소에는 sequence 입력 생성 코드가 아직 없으므로, LSTM 실험은 추가 pipeline 보강이 필요하다.

## 공정성 체크
1. `target_gap > 15 -> delay_risk = 1` 규칙은 tabular 모델과 동일해야 한다.
2. 사용자 split은 `02_train_mlp.ipynb`와 동일해야 한다.
3. decision threshold는 validation set에서 선택하고, test set에서 다시 조정하지 않는다.
