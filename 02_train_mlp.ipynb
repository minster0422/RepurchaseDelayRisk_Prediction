# 02_train_mlp

## 목적
- `DummyClassifier`, `LogisticRegression`, `MLP`를 같은 split에서 비교한다.
- 집계 특징만으로 어느 정도까지 위험 고객을 탐지할 수 있는지 확인한다.

## 입력 파일
- `model_base_hv.csv`
- `01_prepare_data.ipynb`에서 정리한 split 및 라벨 기준

## 출력 파일
- 권장: `results/test_predictions_dummy.csv`
- 권장: `results/test_predictions_logreg.csv`
- 권장: `results/test_predictions_mlp.csv`
- 권장: `results/thresholds.json`

## 현재 상태
- 현재는 **실험 흐름 설명 및 가이드용 노트북** 성격이 강하다.
- 집계 특징 기반 비교 실험의 구조와 공정성 기준을 정리하는 데 초점을 둔다.

## 발표에서 설명할 포인트
- 이 노트북은 `DummyClassifier`, `LogisticRegression`, `MLP`를 같은 조건에서 비교하는 기준선을 설명하는 역할을 맡는다.
- 발표에서는 "집계 특징만으로도 어느 정도 위험 고객 탐지가 가능한가"를 확인하는 단계라고 소개하면 자연스럽다.
- `LogisticRegression`과 `MLP` 모두 validation set에서 threshold tuning을 해야 비교가 공정하다는 점을 함께 강조한다.

## 공정성 체크
1. 세 모델 모두 같은 사용자 split을 사용한다.
2. `LogisticRegression`과 `MLP`는 모두 validation set에서 threshold tuning을 수행한다.
3. test set은 threshold 확정 이후 한 번만 평가한다.
4. accuracy 외에 Recall, F1, ROC-AUC, AP를 함께 기록한다.
