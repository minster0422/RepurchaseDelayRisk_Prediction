# 02_train_mlp

## 목적
- 집계 특징 기반 모델을 한 흐름으로 정리한다.
- `DummyClassifier`, `LogisticRegression`, `MLP`를 같은 split에서 비교한다.
- '집계 특징만으로 충분한가?'라는 질문에 답하는 실험을 담당한다.

## 입력 파일
- `01_prepare_data.ipynb`에서 정의한 tabular feature
- train / val / test split 정보

## 출력 파일
- 권장: `results/test_predictions_dummy.csv`
- 권장: `results/test_predictions_logreg.csv`
- 권장: `results/test_predictions_mlp.csv`
- 권장: `results/thresholds.json`

## 핵심 역할
- `DummyClassifier`로 최소 기준선을 만든다.
- `LogisticRegression`으로 해석 가능한 선형 baseline을 만든다.
- `MLP`로 집계 특징의 비선형 패턴을 학습한다.

## 공정성 체크
1. 세 모델 모두 같은 사용자 split을 사용한다.
2. `LogisticRegression`과 `MLP`는 모두 validation set에서 threshold tuning을 수행한다.
3. test set은 threshold 확정 이후 한 번만 평가한다.
4. accuracy 외에 Recall, F1, ROC-AUC, AP를 함께 기로하다.

## 발표 포인트
- `DummyClassifier`를 이겨야 실험의 최소 설득력이 생긴다.
- `LogisticRegression`은 단순하지만, Recall이 높으면 실무형 기준선으로 의미가 있다.
- `MLP`는 집계 특징만으로도 비선형 관계를 포착할 수 있는지 보여준다.
