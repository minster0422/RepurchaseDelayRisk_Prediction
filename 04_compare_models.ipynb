# 04_compare_models

## 목적
- `DummyClassifier`, `LogisticRegression`, `MLP`, `LSTM`를 한 표에서 공정하게 비교한다.
- threshold tuning, 지표 선택, 발표용 결론 문장을 일관되게 정리한다.

## 입력 파일
- 모델별 test prediction 파일
- 모델별 validation threshold 기록
- `results/model_comparison_template.csv`

## 출력 파일
- 최종 비교 표
- ROC curve
- PR curve
- 발표용 결론 문구

## 현재 상태
- 현재는 **비교 실험 해석 원칙을 정리하는 노트북**이다.
- 실제 수치가 채워지기 전에는 특정 모델을 최종 승자로 단정하지 않는다.

## 발표에서 설명할 포인트
- 이 노트북은 "어떤 기준으로 모델을 비교하고, 어떤 표현까지 말할 수 있는가"를 정리하는 마무리 단계다.
- 발표에서는 accuracy 단독 비교를 피하고 Recall, F1, ROC-AUC, Average Precision, PR curve를 함께 본 이유를 설명하는 데 활용한다.
- 최종 결론은 실제 수치가 채워진 뒤에만 확정하며, 그 전에는 "위험 고객 탐지를 우선할 때 LSTM을 후보로 볼 수 있다" 정도의 신중한 표현이 적절하다.

## 공정성 체크리스트
1. 모든 모델이 같은 사용자 split을 사용했는가?
2. `LogisticRegression`, `MLP`, `LSTM`이 모두 validation set에서 threshold tuning을 했는가?
3. accuracy를 보조 지표로만 쓰고, Recall / F1 / ROC-AUC / AP를 중심으로 정리했는가?
4. test set 성능으로 threshold를 다시 조정하지 않았는가?
5. ROC curve와 함께 PR curve도 확인했는가?

## 결론 원칙
- `MLP`는 집계 특징 기반 분리력이 어느 정도인지 보여주는 모델이다.
- `LSTM`은 순차 정보가 실제로 추가 이득을 주는지 확인하는 모델이다.
- 최종 결론은 실제 결과가 채워진 뒤에만 정리하며, 발표용 문구도 신중하게 쓴다.
