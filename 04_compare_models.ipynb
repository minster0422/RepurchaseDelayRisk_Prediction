# 04_compare_models

## 목적
- `DummyClassifier`, `LogisticRegression`, `MLP`, `LSTM`를 한 표에서 공정하게 비교한다.
- threshold tuning, 지표 선택, 최종 해석 문구를 발표용으로 정리한다.

## 입력 파일
- 모델별 test prediction 파일
- 모델별 validation threshold 기록
- `results/model_comparison_template.csv`

## 출력 파일
- 최종 비교 표
- ROC curve
- PR curve
- 발표용 결론 문구

## 공정성 체크리스트
1. 모든 모델이 같은 사용자 split을 사용했는가?
2. `LogisticRegression`, `MLP`, `LSTM`이 모두 validation set에서 threshold tuning 했는가?
3. accuracy를 보조 지표로만 쓰고, Recall / F1 / ROC-AUC / AP를 중심으로 정리했는가?
4. test set 성능으로 threshold를 다시 조정하지 않았는가?
5. 결과표가 특정 모델에 유리한 순서나 문장으로 편향되지 않았는가?

## 권장 해석 방식
- `MLP`는 ROC-AUC가 높다면 집계 특징 기반 분리력이 강하다고 해석한다.
- `LogisticRegression`은 Recall이 높다면 위험 고객을 넓게 잡는 기준선으로 의미가 있다.
- `LSTM`은 F1-score가 가장 균형적일 때 최종 후보로 검토할 수 있다.

## 결론 문구 템플릿
> 위험 고객 탐지를 우선하는 관점에서 LSTM을 최종 후보로 선정할 수 있다. 다만 집계 특징만으로도 충분한 성능이 나온다면 MLP가 더 단순하고 재현 가능한 대안이 될 수 있다.

## 금지할 표현
- 'LSTM이 압도적으로 최고다'
- '완벽하게 예측한다'
- 'accuracy가 높으므로 가장 좋은 모델이다'
